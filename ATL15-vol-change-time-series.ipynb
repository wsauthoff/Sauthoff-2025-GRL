```python
def ATL15_dhdvdt_time_series(lakename):
    '''
    Create avg. lake dh/dt and dv/dt of all lakes in Siegfried and Fricker, 2018. 
    '''
    # isolate individual lake using gpd buffer
    #lake_gpd = Smith2009_outlines.loc[Smith2009_outlines['Name'] == lakename]
    lake_gpd = SiegfriedFricker2018_outlines.loc[SiegfriedFricker2018_outlines['name'] == lakename]
    lake_buffer = lake_gpd.buffer(1000)
    # define lake bounding box
    x_min = lake_buffer.bounds.values[0,0]; x_max = lake_buffer.bounds.values[0,2]
    y_min = lake_buffer.bounds.values[0,1]; y_max = lake_buffer.bounds.values[0,3]
    # subset ATL15 data set to region of interest
    mask_x = (ATL15_dh.x >= x_min) & (ATL15_dh.x <= x_max)
    mask_y = (ATL15_dh.y >= y_min) & (ATL15_dh.y <= y_max)
    ATL15_dh_sub = ATL15_dh.where(mask_x & mask_y, drop=True)
    # clip ATL15 data to just show (first set crs)
    ATL15_dh_sub.rio.write_crs(3031, inplace=True)
    ATL15_dh_sub_clipped = ATL15_dh_sub.rio.clip(lake_gpd.geometry.values, lake_gpd.crs, drop=False, invert=False)
    # calculate lake volume time series
    IS2_dates = []
    for cyc in range(len(ATL15_dh_sub_clipped.time)-1):
        date_time_str = ATL15_dh_sub_clipped.attrs['rangeBeginningDateTime'][2:10]
        date_time_obj = datetime.datetime.strptime(date_time_str, '%y-%m-%d')
        newdate = date_time_obj + datetime.timedelta(days=ATL15_dh_sub_clipped.time.values[cyc])
        newdate1 = date_time_obj + datetime.timedelta(days=ATL15_dh_sub_clipped.time.values[cyc+1])
        midcycdays = newdate1 - newdate
        midcycdate = newdate + midcycdays
        IS2_dates += [midcycdate]
    IS2_lkavgdhdt = []
    IS2_vols = []
    for cyc in range(len(ATL15_dh_sub_clipped.time)-1):
        arr=ATL15_dh_sub_clipped.delta_h[cyc+1,:,:]-ATL15_dh_sub_clipped.delta_h[cyc,:,:]
        avg_lk_dhdt = np.ma.average(np.ma.MaskedArray(arr, mask=np.isnan(arr)))
        IS2_lkavgdhdt += [avg_lk_dhdt]
        vol = avg_lk_dhdt*SiegfriedFricker2018_outlines[SiegfriedFricker2018_outlines['name'] == lakename]['area (m^2)']
        IS2_vols += [vol.values[0]]
    # plot average lake height time series
    fig, ax = plt.subplots()
    ax.plot(IS2_dates, IS2_lkavgdhdt)
    locator = mdates.AutoDateLocator(minticks=1, maxticks=7)
    formatter = mdates.ConciseDateFormatter(locator)
    ax.xaxis.set_major_locator(locator)
    ax.xaxis.set_major_formatter(formatter)
    ax.set_title('{} average lake height time series'.format(lakename))
    ax.set_xlabel('Year')
    ax.set_ylabel('Avg. height change from previous cycle [m]')
    plt.savefig('/Users/Wilson/Documents/0-code/1-lakeshores/output/IS2-avglkdhdt-lineplots/IS2-avglkdhdt-{}.png'.format(lakename), dpi=300, bbox_inches = "tight")
    plt.close()
    # plot volume change time series
    fig, ax = plt.subplots()
    ax.plot(IS2_dates, np.divide(IS2_vols, 1e+9))
    locator = mdates.AutoDateLocator(minticks=1, maxticks=7)
    formatter = mdates.ConciseDateFormatter(locator)
    ax.xaxis.set_major_locator(locator)
    ax.xaxis.set_major_formatter(formatter)
    ax.set_title('{} volume change time series'.format(lakename))
    ax.set_xlabel('Year')
    ax.set_ylabel('Volume change [km$^3$]')
    plt.savefig('/Users/Wilson/Documents/0-code/1-lakeshores/output/IS2-dvdt-lineplots/IS2-dvdt-{}.png'.format(lakename), dpi=300, bbox_inches = "tight")
    plt.close()
```

```python
def ATL15_dhdvdt_time_series_compare(lakename_S09, lakename_SF18):
    '''
    Compare avg. lake dh/dt and dv/dt of lakes outlines delineated in Smith and others, 2009 
    and later redelineated in Siegfried and Fricker, 2018. 
    '''
    # isolate individual lake using gpd buffer
    lake_S09 = Smith2009_outlines.loc[Smith2009_outlines['Name'] == lakename_S09]
    lake_SF18 = SiegfriedFricker2018_outlines.loc[SiegfriedFricker2018_outlines['name'] == lakename_SF18]
    lake_buffer_S09 = lake_S09.buffer(1000)
    lake_buffer_SF18 = lake_SF18.buffer(1000)
    # define lake bounding box
    x_min = min(lake_buffer_S09.bounds.values[0,0],lake_buffer_SF18.bounds.values[0,0])
    x_max = max(lake_buffer_S09.bounds.values[0,2],lake_buffer_SF18.bounds.values[0,2])
    y_min = min(lake_buffer_S09.bounds.values[0,1],lake_buffer_SF18.bounds.values[0,1])
    y_max = max(lake_buffer_S09.bounds.values[0,3],lake_buffer_SF18.bounds.values[0,3])
    # subset ATL15 data set to region of interest
    mask_x = (ATL15_dh.x >= x_min) & (ATL15_dh.x <= x_max)
    mask_y = (ATL15_dh.y >= y_min) & (ATL15_dh.y <= y_max)
    ATL15_dh_sub = ATL15_dh.where(mask_x & mask_y, drop=True)
    # clip ATL15 data to just show (first set crs)
    ATL15_dh_sub.rio.write_crs(3031, inplace=True)
    ATL15_dh_sub_clipped_S09 = ATL15_dh_sub.rio.clip(lake_S09.geometry.values, lake_S09.crs, drop=False, invert=False)
    ATL15_dh_sub_clipped_SF18 = ATL15_dh_sub.rio.clip(lake_SF18.geometry.values, lake_SF18.crs, drop=False, invert=False)
    # calculate lake volume time series
    IS2_dates = []
    for cyc in range(len(ATL15_dh_sub.time)-1):
        date_time_str = ATL15_dh_sub.attrs['rangeBeginningDateTime'][2:10]
        date_time_obj = datetime.datetime.strptime(date_time_str, '%y-%m-%d')
        newdate = date_time_obj + datetime.timedelta(days=ATL15_dh_sub.time.values[cyc])
        newdate1 = date_time_obj + datetime.timedelta(days=ATL15_dh_sub.time.values[cyc+1])
        midcycdays = newdate1 - newdate
        midcycdate = newdate + midcycdays
        IS2_dates += [midcycdate]
    IS2_lkavgdhdt_S09 = []
    IS2_vols_S09 = []
    IS2_lkavgdhdt_SF18 = []
    IS2_vols_SF18 = []
    for cyc in range(len(ATL15_dh_sub.time)-1):
        arr_S09=ATL15_dh_sub_clipped_S09.delta_h[cyc+1,:,:]-ATL15_dh_sub_clipped_S09.delta_h[cyc,:,:]
        avg_lk_dhdt_S09 = np.ma.average(np.ma.MaskedArray(arr_S09, mask=np.isnan(arr_S09)))
        IS2_lkavgdhdt_S09 += [avg_lk_dhdt_S09]
        vol_S09 = avg_lk_dhdt_S09*Smith2009_outlines[Smith2009_outlines['Name'] == lakename_S09].area
        IS2_vols_S09 += [vol_S09.values[0]]
        arr_SF18=ATL15_dh_sub_clipped_SF18.delta_h[cyc+1,:,:]-ATL15_dh_sub_clipped_SF18.delta_h[cyc,:,:]
        avg_lk_dhdt_SF18 = np.ma.average(np.ma.MaskedArray(arr_SF18, mask=np.isnan(arr_SF18)))
        IS2_lkavgdhdt_SF18 += [avg_lk_dhdt_SF18]
        vol_SF18 = avg_lk_dhdt_SF18*SiegfriedFricker2018_outlines[SiegfriedFricker2018_outlines['name'] == lakename_SF18]['area (m^2)']
        IS2_vols_SF18 += [vol_SF18.values[0]]
    # plot average lake height time series
    fig, ax = plt.subplots()
    ax.plot(IS2_dates, IS2_lkavgdhdt_S09, color='k', linestyle=':')
    ax.plot(IS2_dates, IS2_lkavgdhdt_SF18, color='k', linestyle='--')
    locator = mdates.AutoDateLocator(minticks=1, maxticks=7)
    formatter = mdates.ConciseDateFormatter(locator)
    ax.xaxis.set_major_locator(locator)
    ax.xaxis.set_major_formatter(formatter)
    ax.set_title('{} vs. {} average lake height time series'.format(lakename_S09, lakename_SF18))
    ax.set_xlabel('Year')
    ax.set_ylabel('Avg. height change from previous cycle [m]')
    Smith2009 = plt.Line2D((0, 1), (0, 0), color='k', linestyle=':', linewidth=2)
    SiegfriedFricker2018 = plt.Line2D((0, 1), (0, 0), color='k', linestyle='--', linewidth=2)
    ax.legend([Smith2009,SiegfriedFricker2018],
        ['Smith and others, 2009','Siegfried & Fricker, 2018'], loc='upper right')
    plt.savefig('/Users/Wilson/Documents/0-code/1-lakeshores/output/IS2-outlinecompare-lineplots/IS2-avglkdhdt-{}vs{}.png'\
        .format(lakename_S09, lakename_SF18), dpi=300, bbox_inches = "tight")
    plt.close()
    # plot volume change time series
    fig, ax = plt.subplots()
    ax.plot(IS2_dates, np.divide(IS2_vols_S09, 1e+9), color='k', linestyle=':')
    ax.plot(IS2_dates, np.divide(IS2_vols_SF18, 1e+9), color='k', linestyle='--')
    locator = mdates.AutoDateLocator(minticks=1, maxticks=7)
    formatter = mdates.ConciseDateFormatter(locator)
    ax.xaxis.set_major_locator(locator)
    ax.xaxis.set_major_formatter(formatter)
    ax.set_title('{} vs. {} volume change time series'.format(lakename_S09, lakename_SF18))
    ax.set_xlabel('Year')
    ax.set_ylabel('Volume change [km$^3$]')
    ax.legend([Smith2009,SiegfriedFricker2018],
        ['Smith and others, 2009','Siegfried & Fricker, 2018'], loc='upper right')
    plt.savefig('/Users/Wilson/Documents/0-code/1-lakeshores/output/IS2-outlinecompare-lineplots/IS2-dvdt-{}vs{}.png'\
        .format(lakename_S09, lakename_SF18), dpi=300, bbox_inches = "tight")
    plt.close()
```

```python
def ATL15_dhdvdt_time_series_compare_1to2(lakename_S09, lakename1_SF18, lakename2_SF18):
    '''
    Function to compare S09 and SF18 when 1 lake was converted to 2 lake.
    '''
    # isolate individual lake using gpd buffer
    lake_S09 = Smith2009_outlines.loc[Smith2009_outlines['Name'] == lakename_S09]
    lake1_SF18 = SiegfriedFricker2018_outlines.loc[SiegfriedFricker2018_outlines['name'] == lakename1_SF18]
    lake2_SF18 = SiegfriedFricker2018_outlines.loc[SiegfriedFricker2018_outlines['name'] == lakename2_SF18]
    lake_buffer_S09 = lake1_S09.buffer(1000)
    lake1_buffer_SF18 = lake1_SF18.buffer(1000)
    lake2_buffer_SF18 = lake2_SF18.buffer(1000)
    # define lake bounding box
    x_min = min(lake_buffer_S09.bounds.values[0,0],lake1_buffer_SF18.bounds.values[0,0],lake2_buffer_SF18.bounds.values[0,0])
    x_max = max(lake_buffer_S09.bounds.values[0,2],lake1_buffer_SF18.bounds.values[0,2],lake2_buffer_SF18.bounds.values[0,2])
    y_min = min(lake_buffer_S09.bounds.values[0,1],lake1_buffer_SF18.bounds.values[0,1],lake2_buffer_SF18.bounds.values[0,1])
    y_max = max(lake_buffer_S09.bounds.values[0,3],lake1_buffer_SF18.bounds.values[0,3],lake2_buffer_SF18.bounds.values[0,3])
    # subset ATL15 data set to region of interest
    mask_x = (ATL15_dh.x >= x_min) & (ATL15_dh.x <= x_max)
    mask_y = (ATL15_dh.y >= y_min) & (ATL15_dh.y <= y_max)
    ATL15_dh_sub = ATL15_dh.where(mask_x & mask_y, drop=True)
    # clip ATL15 data to just show (first set crs)
    ATL15_dh_sub.rio.write_crs(3031, inplace=True)
    ATL15_dh_sub_clipped_S09 = ATL15_dh_sub.rio.clip(lake_S09.geometry.values, lake_S09.crs, drop=False, invert=False)
    ATL15_dh_sub_clipped_SF18_lake1 = ATL15_dh_sub.rio.clip(lake1_SF18.geometry.values, lake1_SF18.crs, drop=False, invert=False)
    ATL15_dh_sub_clipped_SF18_lake2 = ATL15_dh_sub.rio.clip(lake2_SF18.geometry.values, lake2_SF18.crs, drop=False, invert=False)
    # calculate lake volume time series
    IS2_dates = []
    for cyc in range(len(ATL15_dh_sub.time)-1):
        date_time_str = ATL15_dh_sub.attrs['rangeBeginningDateTime'][2:10]
        date_time_obj = datetime.datetime.strptime(date_time_str, '%y-%m-%d')
        newdate = date_time_obj + datetime.timedelta(days=ATL15_dh_sub.time.values[cyc])
        newdate1 = date_time_obj + datetime.timedelta(days=ATL15_dh_sub.time.values[cyc+1])
        midcycdays = newdate1 - newdate
        midcycdate = newdate + midcycdays
        IS2_dates += [midcycdate]
    IS2_lkavgdhdt_S09= []
    IS2_vols_S09 = []
    IS2_lkavgdhdt_SF18_lake1 = []
    IS2_vols_SF18_lake1 = []
    IS2_lkavgdhdt_SF18_lake2 = []
    IS2_vols_SF18_lake2 = []
    for cyc in range(len(ATL15_dh_sub.time)-1):
        # S09 lake
        arr_S09 = ATL15_dh_sub_clipped_S09.delta_h[cyc+1,:,:]-ATL15_dh_sub_clipped_S09.delta_h[cyc,:,:]
        avg_lk_dhdt_S09 = np.ma.average(np.ma.MaskedArray(arr_S09, mask=np.isnan(arr_S09)))
        IS2_lkavgdhdt_S09 += [avg_lk_dhdt_S09]
        vol_S09 = avg_lk_dhdt_S09*Smith2009_outlines[Smith2009_outlines['Name'] == lakename_S09].area
        IS2_vols_S09 += [vol_S09.values[0]]
        # SF18 lake 1
        arr_SF18_lake1 = ATL15_dh_sub_clipped_SF18_lake1.delta_h[cyc+1,:,:]-ATL15_dh_sub_clipped_SF18_lake1.delta_h[cyc,:,:]
        avg_lk_dhdt_SF18_lake1 = np.ma.average(np.ma.MaskedArray(arr_SF18_lake1, mask=np.isnan(arr_SF18_lake1)))
        IS2_lkavgdhdt_SF18_lake1 += [avg_lk_dhdt_SF18_lake1]
        vol_SF18_lake1 = avg_lk_dhdt_SF18_lake1*SiegfriedFricker2018_outlines[SiegfriedFricker2018_outlines['name'] == lakename1_SF18]['area (m^2)']
        IS2_vols_SF18_lake1 += [vol_SF18_lake1.values[0]]
        # SF18 lake 2
        arr_SF18_lake2 = ATL15_dh_sub_clipped_SF18_lake2.delta_h[cyc+1,:,:]-ATL15_dh_sub_clipped_SF18_lake2.delta_h[cyc,:,:]
        avg_lk_dhdt_SF18_lake2 = np.ma.average(np.ma.MaskedArray(arr_SF18_lake2, mask=np.isnan(arr_SF18_lake2)))
        IS2_lkavgdhdt_SF18_lake2 += [avg_lk_dhdt_SF18_lake2]
        vol_SF18_lake2 = avg_lk_dhdt_SF18_lake2*SiegfriedFricker2018_outlines[SiegfriedFricker2018_outlines['name'] == lakename2_SF18]['area (m^2)']
        IS2_vols_SF18_lake2 += [vol_SF18_lake2.values[0]]
    # plot average lake height time series
    fig, ax = plt.subplots()
    # ax.plot(IS2_dates, IS2_lkavgdhdt_S09_lake1+IS2_lkavgdhdt_S09_lake2, color='k', linestyle=':')
    ax.plot(IS2_dates, IS2_lkavgdhdt_S09, color='k', linestyle=':')
    ax.plot(IS2_dates, [sum(x) for x in zip(IS2_lkavgdhdt_SF18_lake1, IS2_lkavgdhdt_SF18_lake2)], color='k', linestyle='--')
    locator = mdates.AutoDateLocator(minticks=1, maxticks=7)
    formatter = mdates.ConciseDateFormatter(locator)
    ax.xaxis.set_major_locator(locator)
    ax.xaxis.set_major_formatter(formatter)
    ax.set_title('{} vs. {}+{} average lake height time series'.format(lakename_S09, lakename1_SF18, lakename2_SF18))
    ax.set_xlabel('Year')
    ax.set_ylabel('Avg. height change from previous cycle [m]')
    Smith2009 = plt.Line2D((0, 1), (0, 0), color='k', linestyle=':', linewidth=2)
    SiegfriedFricker2018 = plt.Line2D((0, 1), (0, 0), color='k', linestyle='--', linewidth=2)
    ax.legend([Smith2009,SiegfriedFricker2018],
        ['Smith and others, 2009','Siegfried & Fricker, 2018'], loc='upper right')
    plt.savefig('/Users/Wilson/Documents/0-code/1-lakeshores/output/IS2-outlinecompare-lineplots/IS2-avglkdhdt-{}vs{}+{}.png'\
        .format(lakename_S09, lakename1_SF18, lakename2_SF18), dpi=300, bbox_inches = "tight")
    plt.close()
    # plot volume change time series
    fig, ax = plt.subplots()
    # ax.plot(IS2_dates, np.divide(IS2_vols_S09_lake1+IS2_vols_S09_lake2, 1e+9), color='k', linestyle=':')
    ax.plot(IS2_dates, np.divide(IS2_vols_S09, 1e+9), color='k', linestyle=':')
    ax.plot(IS2_dates, np.divide([sum(x) for x in zip(IS2_vols_SF18_lake1, IS2_vols_SF18_lake2)], 1e+9), color='k', linestyle='--')
    locator = mdates.AutoDateLocator(minticks=1, maxticks=7)
    formatter = mdates.ConciseDateFormatter(locator)
    ax.xaxis.set_major_locator(locator)
    ax.xaxis.set_major_formatter(formatter)
    ax.set_title('{} vs. {}+{} volume change time series'.format(lakename_S09, lakename1_SF18, lakename2_SF18))
    ax.set_xlabel('Year')
    ax.set_ylabel('Volume change [km$^3$]')
    ax.legend([Smith2009,SiegfriedFricker2018],
        ['Smith and others, 2009','Siegfried & Fricker, 2018'], loc='upper right')
    plt.savefig('/Users/Wilson/Documents/0-code/1-lakeshores/output/IS2-outlinecompare-lineplots/IS2-dvdt-{}vs{}+{}.png'\
        .format(lakename_S09, lakename1_SF18, lakename2_SF18), dpi=300, bbox_inches = "tight")
    plt.close()
```
